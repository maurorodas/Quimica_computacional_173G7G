## üìö Introducci√≥n a la Mec√°nica Molecular

### ¬øQu√© es la Mec√°nica Molecular?

La **mec√°nica molecular (MM)** es un m√©todo computacional que trata a las mol√©culas como sistemas de √°tomos unidos por enlaces, utilizando las leyes de la mec√°nica cl√°sica (newtoniana) en lugar de la mec√°nica cu√°ntica.

> "En mec√°nica molecular, los √°tomos son esferas conectadas por resortes" - Andrew R. Leach

### Diferencias Fundamentales: MM vs QM

| Aspecto | Mec√°nica Molecular (MM) | Mec√°nica Cu√°ntica (QM) |

| **Tratamiento** | √Åtomos como part√≠culas | Electrones expl√≠citos |

| **Reacciones qu√≠micas** | ‚ùå No | ‚úÖ S√≠ |

## ‚ö° Aproximaciones Fundamentales

### 1. Aproximaci√≥n de Born-Oppenheimer

La mec√°nica molecular asume que:
- Los **n√∫cleos** se mueven mucho m√°s lento que los **electrones**
- Los electrones se ajustan instant√°neamente a las posiciones nucleares
- Podemos separar el movimiento nuclear del electr√≥nico

$$\Psi_{\text{total}} = \psi_{\text{electrones}}(\mathbf{r}; \mathbf{R}) \cdot \psi_{\text{n√∫cleos}}(\mathbf{R})$$

Donde:
- $\mathbf{r}$ = coordenadas electr√≥nicas
- $\mathbf{R}$ = coordenadas nucleares

### 2. Superficie de Energ√≠a Potencial (PES)

La energ√≠a total del sistema depende √∫nicamente de las **posiciones at√≥micas**:

$$E = E(\mathbf{R}_1, \mathbf{R}_2, ..., \mathbf{R}_N)$$

Esta superficie de energ√≠a potencial (PES) describe:
- M√≠nimos = estructuras estables (conformaciones)
- Puntos de silla = estados de transici√≥n (caminos de reacci√≥n)
- M√°ximos = barreras energ√©ticas

## üî¨ El Campo de Fuerza (Force Field)

### Definici√≥n

Un **campo de fuerza** es un conjunto de:
1. **Funciones matem√°ticas** que describen c√≥mo cambia la energ√≠a con la geometr√≠a
2. **Par√°metros** (constantes) espec√≠ficos para diferentes tipos de √°tomos

### Ecuaci√≥n General de Energ√≠a

La energ√≠a total se calcula como la suma de m√∫ltiples t√©rminos:

$$E_{\text{total}} = E_{\text{enlace}} + E_{\text{√°ngulo}} + E_{\text{torsi√≥n}} + E_{\text{vdW}} + E_{\text{elec}} + E_{\text{cruzados}}$$

Cada t√©rmino representa un tipo espec√≠fico de interacci√≥n molecular.

## üìê T√©rminos Enlazados (Bonded Terms)

### 1. Energ√≠a de Estiramiento de Enlaces

Modela el cambio de energ√≠a cuando un enlace se estira o comprime respecto a su longitud de equilibrio.

#### Aproximaci√≥n Arm√≥nica (Ley de Hooke)

$$E_{\text{enlace}} = \sum_{\text{enlaces}} \frac{1}{2} k_b (r - r_0)^2$$

Donde:
- $k_b$ = constante de fuerza del enlace (kcal/mol/≈≤ o kJ/mol/nm√Ç¬≤)
- $r$ = longitud actual del enlace
- $r_0$ = longitud de equilibrio del enlace

**Caracter√≠sticas:**
- ‚úÖ Simple y r√°pida de calcular
- ‚úÖ V√°lida cerca del equilibrio
- ‚ùå No describe ruptura de enlaces
- ‚ùå Falla en deformaciones grandes

#### Potencial de Morse (M√°s Realista)

$$E_{\text{Morse}} = D_e \left[1 - e^{-a(r-r_0)}\right]^2$$

Donde:
- $D_e$ = profundidad del pozo (energ√≠a de disociaci√≥n)
- $a$ = constante que controla el ancho del pozo
- Permite modelar la disociaci√≥n del enlace

- $\theta$ = √°ngulo actual (en radianes)
- $\theta_0$ = √°ngulo de equilibrio

### 3. Energ√≠a Torsional (√Ångulos Diedros)

Describe la barrera energ√©tica asociada con la rotaci√≥n alrededor de enlaces simples.

#### Serie de Fourier

$$E_{\text{torsi√≥n}} = \sum_{\text{torsiones}} \sum_{n} \frac{V_n}{2} \left[1 + \cos(n\phi - \gamma_n)\right]$$

Donde:
- $V_n$ = amplitud de la barrera para la componente n
- $n$ = multiplicidad (periodicidad de la funci√≥n)
- $\phi$ = √°ngulo diedro actual
- $\gamma_n$ = fase (√°ngulo de offset)

**Casos comunes:**

**Etano (C-C):** Barrera de rotaci√≥n ~3 kcal/mol

**Butano (C-C-C-C):** M√∫ltiples conformaciones (anti, gauche)

## üåê T√©rminos No Enlazados (Non-bonded Terms)

### 1. Interacciones de Van der Waals

Las interacciones de Van der Waals incluyen:
- Fuerzas de dispersi√≥n de London (dipolo inducido)
- Repulsi√≥n de Pauli (overlapping de nubes electr√≥nicas)

#### Potencial de Lennard-Jones (12-6)

El m√°s utilizado en campos de fuerza:

$$E_{\text{vdW}} = \sum_{i<j} 4\epsilon_{ij} \left[\left(\frac{\sigma_{ij}}{r_{ij}}\right)^{12} - \left(\frac{\sigma_{ij}}{r_{ij}}\right)^{6}\right]$$

Donde:
- $\epsilon_{ij}$ = profundidad del pozo potencial (energ√≠a de interacci√≥n m√≠nima)
- $\sigma_{ij}$ = distancia a la cual la energ√≠a es cero (di√°metro de colisi√≥n)
- $r_{ij}$ = distancia entre √°tomos i y j

**Componentes:**
- T√©rmino $r^{-12}$: **Repulsi√≥n** a corta distancia (Pauli)
- T√©rmino $r^{-6}$: **Atracci√≥n** a larga distancia (London)

#### Reglas de Combinaci√≥n

Para pares de √°tomos diferentes (i ‚â† j):

**Lorentz-Berthelot (m√°s com√∫n):**
$$\sigma_{ij} = \frac{\sigma_i + \sigma_j}{2}$$
$$\epsilon_{ij} = \sqrt{\epsilon_i \cdot \epsilon_j}$$

**Geometric mean:**
$$\sigma_{ij} = \sqrt{\sigma_i \cdot \sigma_j}$$

**Referencias:**
- **Lennard-Jones, J. E.** (1924). "On the determination of molecular fields." *Proc. R. Soc. Lond. A* **106**(738), 463-477. DOI: [10.1098/rspa.1924.0082](https://doi.org/10.1098/rspa.1924.0082)

### 2. Interacciones Electrost√°ticas

Describen las interacciones entre cargas parciales en los √°tomos.

#### Ley de Coulomb

$$E_{\text{elec}} = \sum_{i<j} \frac{q_i q_j}{4\pi\epsilon_0 r_{ij}}$$

En unidades pr√°cticas:

$$E_{\text{elec}} = 332.06 \sum_{i<j} \frac{q_i q_j}{r_{ij}} \quad \text{(kcal/mol, con r en √Ö)}$$

Donde:
- $q_i, q_j$ = cargas parciales at√≥micas (en unidades de carga elemental e)
- $\epsilon_0$ = permitividad del vac√≠o
- $r_{ij}$ = distancia entre √°tomos i y j
- 332.06 = factor de conversi√≥n de unidades

**Caracter√≠sticas:**
- Decae lentamente ($r^{-1}$) ‚Üí interacciones a largo alcance
- Dominante en mol√©culas polares y sistemas i√≥nicos
- Requiere tratamiento especial en sistemas peri√≥dicos (Ewald summation)

### Cargas Parciales At√≥micas

Las cargas parciales se pueden obtener mediante:

1. **M√©todos basados en QM:**
   - Mulliken
   - Natural Population Analysis (NPA)
   - RESP (Restrained Electrostatic Potential)
   - CM5 (Charge Model 5)

2. **M√©todos emp√≠ricos:**
   - Gasteiger charges
   - MMFF94 charges
   - QEq (Charge Equilibration)

**Restricci√≥n fundamental:**
$$\sum_i q_i = Q_{\text{total}}$$

(La suma de cargas parciales debe igualar la carga total de la mol√©cula)

**Referencias:**
- **Bayly, C. I. et al.** (1993). "A well-behaved electrostatic potential based method using charge restraints." *J. Phys. Chem.* **97**, 10269-10280. DOI: [10.1021/j100142a004](https://doi.org/10.1021/j100142a004)

## üîÄ T√©rminos Cruzados (Cross Terms)

Algunos campos de fuerza avanzados incluyen **acoplamientos** entre diferentes tipos de deformaci√≥n:

### Estiramiento-Estiramiento

$$E_{ss} = k_{ss}(r_1 - r_1^0)(r_2 - r_2^0)$$

### Estiramiento-Flexi√≥n

$$E_{sb} = k_{sb}(r - r_0)(\theta - \theta_0)$$

### Flexi√≥n-Flexi√≥n

$$E_{bb} = k_{bb}(\theta_1 - \theta_1^0)(\theta_2 - \theta_2^0)$$

**Ejemplos de campos de fuerza con t√©rminos cruzados:**

- CFF (Consistent Force Field)
- COMPASS (Condensed-phase Optimized Molecular Potentials for Atomistic Simulation Studies)

**Ventajas:**
- ‚úÖ Mayor precisi√≥n en frecuencias vibracionales
- ‚úÖ Mejor reproducci√≥n de propiedades termodin√°micas

**Desventajas:**
- ‚ùå Mayor costo computacional
- ‚ùå M√°s par√°metros para ajustar
- ‚ùå Mayor complejidad en la parametrizaci√≥n

### UFF (Universal Force Field)

**Caracter√≠sticas:**

**Aplicaciones:**

- ‚úÖ Complejos prote√≠na-ligando
- ‚úÖ Din√°mica molecular en agua
