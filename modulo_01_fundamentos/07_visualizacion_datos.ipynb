### Primer GrÃ¡fico Simple


````python
# Datos simples
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear grÃ¡fico
plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Mi Primer GrÃ¡fico')
plt.grid(True, alpha=0.3)
plt.show()
````

### Dos Enfoques: pyplot vs Orientado a Objetos


````python
# Enfoque 1: pyplot (estilo MATLAB)
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('Estilo pyplot')

# Enfoque 2: Orientado a Objetos (recomendado)
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[1].plot(x, y)
ax[1].set_title('Estilo OO (recomendado)')

plt.tight_layout()
plt.show()

print("ğŸ’¡ El enfoque OO da mayor control y es mejor para grÃ¡ficos complejos")
````

### GrÃ¡ficos de LÃ­neas


````python
# Datos: EnergÃ­a vs Distancia (curva de Morse)
# E(r) = D_e * (1 - exp(-a(r - r_e)))^2

r = np.linspace(0.5, 4, 200)  # Distancia (Ã…)
D_e = 100  # EnergÃ­a de disociaciÃ³n (kcal/mol)
r_e = 1.5  # Distancia de equilibrio (Ã…)
a = 1.5    # ParÃ¡metro de anchura

E_morse = D_e * (1 - np.exp(-a * (r - r_e)))**2

# Crear grÃ¡fico
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(r, E_morse, 'b-', linewidth=2, label='Potencial de Morse')
ax.axhline(y=D_e, color='r', linestyle='--', alpha=0.5, label=f'$D_e$ = {D_e} kcal/mol')
ax.axvline(x=r_e, color='g', linestyle='--', alpha=0.5, label=f'$r_e$ = {r_e} Ã…')

ax.set_xlabel('Distancia de enlace (Ã…)', fontsize=12)
ax.set_ylabel('EnergÃ­a (kcal/mol)', fontsize=12)
ax.set_title('Curva de Potencial de Morse', fontsize=14, fontweight='bold')
ax.legend(loc='best')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

### MÃºltiples LÃ­neas y Estilos


````python
# EnergÃ­as de conformeros
angulo = np.linspace(0, 360, 100)

# Diferentes barreras rotacionales
E_etano = 3 * (1 - np.cos(3 * np.radians(angulo))) / 2  # Eclipsada/alternada
E_butano = 5 * (1 - np.cos(3 * np.radians(angulo))) / 2
E_propano = 3.5 * (1 - np.cos(3 * np.radians(angulo))) / 2

fig, ax = plt.subplots(figsize=(10, 6))

# Diferentes estilos de lÃ­nea
ax.plot(angulo, E_etano, 'b-', linewidth=2, label='Etano')
ax.plot(angulo, E_propano, 'g--', linewidth=2, label='Propano')
ax.plot(angulo, E_butano, 'r:', linewidth=2.5, label='Butano')

ax.set_xlabel('Ãngulo diedro (Â°)', fontsize=12)
ax.set_ylabel('EnergÃ­a relativa (kcal/mol)', fontsize=12)
ax.set_title('Perfil de EnergÃ­a Rotacional', fontsize=14, fontweight='bold')
ax.set_xlim(0, 360)
ax.set_xticks([0, 60, 120, 180, 240, 300, 360])
ax.legend(loc='upper right', fontsize=11)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

### GrÃ¡ficos de DispersiÃ³n (Scatter)


````python
# Generar datos simulados: CorrelaciÃ³n estructura-propiedad
np.random.seed(42)
n_moleculas = 50

# Masa molecular vs Punto de ebulliciÃ³n
masa_mol = np.random.uniform(20, 200, n_moleculas)
punto_eb = 50 + 1.2 * masa_mol + np.random.normal(0, 20, n_moleculas)

# Clasificar por familia quÃ­mica
familia = np.random.choice(['Alcano', 'Alcohol', 'Ã‰ster'], n_moleculas)
colores = {'Alcano': 'blue', 'Alcohol': 'red', 'Ã‰ster': 'green'}

fig, ax = plt.subplots(figsize=(9, 6))

# Graficar por familia
for fam in ['Alcano', 'Alcohol', 'Ã‰ster']:
    mask = familia == fam
    ax.scatter(masa_mol[mask], punto_eb[mask], 
              c=colores[fam], label=fam, alpha=0.6, s=80, edgecolors='black')

# LÃ­nea de tendencia
z = np.polyfit(masa_mol, punto_eb, 1)
p = np.poly1d(z)
ax.plot(masa_mol, p(masa_mol), "k--", alpha=0.5, label=f'Ajuste: y={z[0]:.2f}x+{z[1]:.1f}')

ax.set_xlabel('Masa molecular (g/mol)', fontsize=12)
ax.set_ylabel('Punto de ebulliciÃ³n (Â°C)', fontsize=12)
ax.set_title('CorrelaciÃ³n: Masa Molecular vs Punto de EbulliciÃ³n', fontsize=13, fontweight='bold')
ax.legend(loc='upper left')
ax.grid(True, alpha=0.3)

# Calcular RÂ²
from scipy.stats import pearsonr
r, p_value = pearsonr(masa_mol, punto_eb)
ax.text(0.05, 0.95, f'RÂ² = {r**2:.3f}\np = {p_value:.2e}', 
        transform=ax.transAxes, verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.show()
````

### GrÃ¡ficos de Barras


````python
# ComparaciÃ³n de energÃ­as relativas de conformeros
conformeros = ['Anti', 'Gauche+', 'Gauche-', 'Eclipsado']
energias = [0.0, 0.8, 0.8, 3.2]  # kcal/mol
colores_barras = ['green', 'orange', 'orange', 'red']

fig, ax = plt.subplots(figsize=(8, 6))

bars = ax.bar(conformeros, energias, color=colores_barras, alpha=0.7, edgecolor='black', linewidth=1.5)

# Agregar valores sobre las barras
for bar, energia in zip(bars, energias):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{energia:.1f}',
            ha='center', va='bottom', fontsize=11, fontweight='bold')

ax.set_ylabel('EnergÃ­a relativa (kcal/mol)', fontsize=12)
ax.set_title('EnergÃ­as de Conformeros de Butano', fontsize=14, fontweight='bold')
ax.set_ylim(0, max(energias) * 1.2)
ax.axhline(y=1.5, color='gray', linestyle='--', alpha=0.3, label='Barrera tÃ­pica')
ax.legend()
ax.grid(True, axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
````

### Barras Agrupadas


````python
# ComparaciÃ³n de propiedades calculadas con diferentes mÃ©todos
metodos = ['HF', 'B3LYP', 'M06-2X', 'MP2']
energia = [-154.2, -155.8, -155.5, -155.3]  # hartrees
momento_dipolar = [1.8, 1.9, 1.85, 1.88]  # Debye

x = np.arange(len(metodos))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))

bars1 = ax.bar(x - width/2, energia, width, label='EnergÃ­a (Ha)', color='steelblue', alpha=0.8)
ax2 = ax.twinx()
bars2 = ax2.bar(x + width/2, momento_dipolar, width, label='Momento dipolar (D)', color='coral', alpha=0.8)

ax.set_xlabel('MÃ©todo computacional', fontsize=12)
ax.set_ylabel('EnergÃ­a (Hartrees)', fontsize=12, color='steelblue')
ax2.set_ylabel('Momento dipolar (Debye)', fontsize=12, color='coral')
ax.set_title('ComparaciÃ³n de MÃ©todos Computacionales - MolÃ©cula de Agua', fontsize=13, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(metodos)
ax.tick_params(axis='y', labelcolor='steelblue')
ax2.tick_params(axis='y', labelcolor='coral')

# Combinar leyendas
lines1, labels1 = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

ax.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
````

### Histogramas


````python
# DistribuciÃ³n de distancias interatÃ³micas en una trayectoria MD
np.random.seed(10)
distancias = np.random.normal(1.5, 0.2, 1000)  # Media 1.5 Ã…, std 0.2 Ã…

fig, ax = plt.subplots(figsize=(9, 6))

n, bins, patches = ax.hist(distancias, bins=30, color='skyblue', 
                           edgecolor='black', alpha=0.7, density=True)

# Agregar curva normal ajustada
mu, sigma = distancias.mean(), distancias.std()
x_fit = np.linspace(distancias.min(), distancias.max(), 100)
y_fit = (1/(sigma * np.sqrt(2*np.pi))) * np.exp(-0.5*((x_fit-mu)/sigma)**2)
ax.plot(x_fit, y_fit, 'r-', linewidth=2, label=f'Normal(Î¼={mu:.2f}, Ïƒ={sigma:.2f})')

# LÃ­nea vertical en la media
ax.axvline(mu, color='red', linestyle='--', linewidth=2, alpha=0.7)

ax.set_xlabel('Distancia O-H (Ã…)', fontsize=12)
ax.set_ylabel('Densidad de probabilidad', fontsize=12)
ax.set_title('DistribuciÃ³n de Distancias de Enlace - Trayectoria MD', fontsize=13, fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

### Subplots (MÃºltiples GrÃ¡ficos)


````python
# AnÃ¡lisis de conformeros: mÃºltiples propiedades
np.random.seed(42)
n_conf = 30

data = {
    'EnergÃ­a': np.random.exponential(2, n_conf),
    'Dipolo': np.random.gamma(2, 1, n_conf),
    'HOMO': np.random.normal(-5.5, 0.5, n_conf),
    'LUMO': np.random.normal(-1.5, 0.3, n_conf)
}

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('AnÃ¡lisis de Propiedades de Conformeros', fontsize=16, fontweight='bold')

# EnergÃ­a vs Ãndice
axes[0, 0].scatter(range(n_conf), data['EnergÃ­a'], c=data['EnergÃ­a'], 
                   cmap='RdYlGn_r', s=100, edgecolors='black')
axes[0, 0].set_xlabel('Conformero ID')
axes[0, 0].set_ylabel('EnergÃ­a relativa (kcal/mol)')
axes[0, 0].set_title('EnergÃ­as Relativas')
axes[0, 0].grid(True, alpha=0.3)

# Histograma de Momento Dipolar
axes[0, 1].hist(data['Dipolo'], bins=15, color='lightcoral', edgecolor='black', alpha=0.7)
axes[0, 1].axvline(np.mean(data['Dipolo']), color='red', linestyle='--', linewidth=2)
axes[0, 1].set_xlabel('Momento dipolar (Debye)')
axes[0, 1].set_ylabel('Frecuencia')
axes[0, 1].set_title('DistribuciÃ³n de Momentos Dipolares')
axes[0, 1].grid(True, alpha=0.3)

# Gap HOMO-LUMO
gap = np.array(data['LUMO']) - np.array(data['HOMO'])
axes[1, 0].scatter(data['EnergÃ­a'], gap, s=100, alpha=0.6, c='purple', edgecolors='black')
axes[1, 0].set_xlabel('EnergÃ­a relativa (kcal/mol)')
axes[1, 0].set_ylabel('Gap HOMO-LUMO (eV)')
axes[1, 0].set_title('EnergÃ­a vs Gap')
axes[1, 0].grid(True, alpha=0.3)

# Boxplot de HOMO y LUMO
axes[1, 1].boxplot([data['HOMO'], data['LUMO']], labels=['HOMO', 'LUMO'],
                   patch_artist=True, 
                   boxprops=dict(facecolor='lightblue', alpha=0.7))
axes[1, 1].set_ylabel('EnergÃ­a orbital (eV)')
axes[1, 1].set_title('DistribuciÃ³n de EnergÃ­as Orbitales')
axes[1, 1].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
````

### Heatmaps (Mapas de Calor)


````python
# Matriz de correlaciÃ³n de propiedades moleculares
np.random.seed(20)

propiedades = ['Masa', 'LogP', 'TPSA', 'HBD', 'HBA', 'Rotatable']
n_mols = 100

datos_moleculas = pd.DataFrame({
    'Masa': np.random.normal(200, 50, n_mols),
    'LogP': np.random.normal(2, 1, n_mols),
    'TPSA': np.random.normal(80, 30, n_mols),
    'HBD': np.random.poisson(2, n_mols),
    'HBA': np.random.poisson(4, n_mols),
    'Rotatable': np.random.poisson(5, n_mols)
})

# Calcular matriz de correlaciÃ³n
corr_matrix = datos_moleculas.corr()

fig, ax = plt.subplots(figsize=(10, 8))

im = ax.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1, aspect='auto')

# Configurar ticks
ax.set_xticks(np.arange(len(propiedades)))
ax.set_yticks(np.arange(len(propiedades)))
ax.set_xticklabels(propiedades)
ax.set_yticklabels(propiedades)

# Rotar labels del eje x
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Agregar valores en las celdas
for i in range(len(propiedades)):
    for j in range(len(propiedades)):
        text = ax.text(j, i, f'{corr_matrix.iloc[i, j]:.2f}',
                      ha="center", va="center", color="black", fontsize=10)

ax.set_title('Matriz de CorrelaciÃ³n de Propiedades Moleculares', fontsize=14, fontweight='bold', pad=20)

# Colorbar
cbar = plt.colorbar(im, ax=ax)
cbar.set_label('Coeficiente de correlaciÃ³n', rotation=270, labelpad=20)

plt.tight_layout()
plt.show()
````

### GrÃ¡ficos de Contorno (Superficies de EnergÃ­a)


````python
# Superficie de EnergÃ­a Potencial 2D
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

# FunciÃ³n de ejemplo: doble pozo
Z = (X**2 - 1)**2 + (Y**2 - 1)**2 + 0.1*X*Y

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Contornos rellenados
contourf = ax1.contourf(X, Y, Z, levels=20, cmap='viridis')
ax1.contour(X, Y, Z, levels=10, colors='white', linewidths=0.5, alpha=0.4)
ax1.set_xlabel('Coordenada de reacciÃ³n 1 (Ã…)', fontsize=11)
ax1.set_ylabel('Coordenada de reacciÃ³n 2 (Ã…)', fontsize=11)
ax1.set_title('Superficie de EnergÃ­a Potencial (Contornos)', fontsize=13, fontweight='bold')
plt.colorbar(contourf, ax=ax1, label='EnergÃ­a (kcal/mol)')
ax1.plot([-1, 1], [-1, 1], 'r*', markersize=15, label='MÃ­nimos')
ax1.legend()

# Contornos de lÃ­nea
contour = ax2.contour(X, Y, Z, levels=15, cmap='plasma')
ax2.clabel(contour, inline=True, fontsize=8)
ax2.set_xlabel('Coordenada de reacciÃ³n 1 (Ã…)', fontsize=11)
ax2.set_ylabel('Coordenada de reacciÃ³n 2 (Ã…)', fontsize=11)
ax2.set_title('PES con Etiquetas de EnergÃ­a', fontsize=13, fontweight='bold')
ax2.plot([-1, 1], [-1, 1], 'r*', markersize=15)

plt.tight_layout()
plt.show()
````

### GrÃ¡ficos 3D


````python
from mpl_toolkits.mplot3d import Axes3D

# Superficie 3D de energÃ­a
fig = plt.figure(figsize=(12, 5))

# Superficie
ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X, Y, Z, cmap='coolwarm', alpha=0.8, edgecolor='none')
ax1.set_xlabel('X (Ã…)', fontsize=10)
ax1.set_ylabel('Y (Ã…)', fontsize=10)
ax1.set_zlabel('EnergÃ­a (kcal/mol)', fontsize=10)
ax1.set_title('Superficie de EnergÃ­a Potencial 3D', fontsize=12, fontweight='bold')
fig.colorbar(surf, ax=ax1, shrink=0.5)

# Wireframe
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_wireframe(X, Y, Z, color='blue', alpha=0.3, linewidth=0.5)
ax2.contour(X, Y, Z, zdir='z', offset=Z.min(), cmap='viridis', levels=10, alpha=0.5)
ax2.set_xlabel('X (Ã…)', fontsize=10)
ax2.set_ylabel('Y (Ã…)', fontsize=10)
ax2.set_zlabel('EnergÃ­a (kcal/mol)', fontsize=10)
ax2.set_title('Wireframe con ProyecciÃ³n', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
````

### Scatter Plot Interactivo


````python
# Datos de ejemplo: Descriptores moleculares
np.random.seed(42)
n = 100

df_mol = pd.DataFrame({
    'Masa_Molecular': np.random.uniform(100, 500, n),
    'LogP': np.random.uniform(-2, 6, n),
    'TPSA': np.random.uniform(20, 150, n),
    'Actividad': np.random.choice(['Alta', 'Media', 'Baja'], n),
    'Nombre': [f'Mol_{i:03d}' for i in range(n)]
})

# Crear grÃ¡fico interactivo
fig = px.scatter(df_mol, 
                 x='Masa_Molecular', 
                 y='LogP', 
                 size='TPSA',
                 color='Actividad',
                 hover_data=['Nombre', 'TPSA'],
                 title='Espacio QuÃ­mico de MolÃ©culas (Interactivo)',
                 labels={'Masa_Molecular': 'Masa Molecular (g/mol)',
                        'LogP': 'LogP (coef. particiÃ³n)'},
                 color_discrete_map={'Alta': 'red', 'Media': 'yellow', 'Baja': 'blue'})

fig.update_layout(
    font=dict(size=12),
    title_font_size=16,
    hoverlabel=dict(bgcolor="white"),
    height=600
)

fig.show()

print("ğŸ’¡ InteracciÃ³n: Pasa el mouse sobre los puntos para ver detalles")
print("   Zoom: Selecciona Ã¡rea con el mouse")
print("   Pan: MantÃ©n y arrastra")
````

### GrÃ¡fico 3D Interactivo


````python
# Superficie 3D interactiva
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y, colorscale='Viridis')])

fig.update_layout(
    title='Superficie de EnergÃ­a Potencial Interactiva',
    scene=dict(
        xaxis_title='Coordenada X (Ã…)',
        yaxis_title='Coordenada Y (Ã…)',
        zaxis_title='EnergÃ­a (kcal/mol)'
    ),
    height=600
)

fig.show()

print("ğŸ® RotaciÃ³n: Arrastra con el mouse")
print("   Zoom: Scroll o pinch")
````

### GrÃ¡ficos de LÃ­neas Interactivos


````python
# Trayectoria de dinÃ¡mica molecular
tiempo = np.linspace(0, 100, 1000)  # ps
rmsd = 2 + 0.5*np.sin(tiempo/10) + np.random.normal(0, 0.2, 1000)
rg = 15 + 0.3*np.cos(tiempo/8) + np.random.normal(0, 0.15, 1000)

fig = make_subplots(rows=2, cols=1,
                    subplot_titles=('RMSD vs Tiempo', 'Radio de Giro vs Tiempo'))

fig.add_trace(
    go.Scatter(x=tiempo, y=rmsd, mode='lines', name='RMSD',
              line=dict(color='blue', width=1.5)),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(x=tiempo, y=rg, mode='lines', name='Radio de Giro',
              line=dict(color='red', width=1.5)),
    row=2, col=1
)

fig.update_xaxes(title_text="Tiempo (ps)", row=2, col=1)
fig.update_yaxes(title_text="RMSD (Ã…)", row=1, col=1)
fig.update_yaxes(title_text="Rg (Ã…)", row=2, col=1)

fig.update_layout(height=700, title_text="AnÃ¡lisis de Trayectoria MD", showlegend=False)

fig.show()

print("ğŸ’¡ Haz doble clic en un grÃ¡fico para hacer zoom completo")
````

### Heatmap Interactivo


````python
# Matriz de distancias interatÃ³micas
atomos = ['C1', 'C2', 'C3', 'O1', 'N1', 'H1', 'H2']
n_atomos = len(atomos)

# Generar matriz de distancias simulada (simÃ©trica)
np.random.seed(10)
matriz_dist = np.random.uniform(1, 5, (n_atomos, n_atomos))
matriz_dist = (matriz_dist + matriz_dist.T) / 2
np.fill_diagonal(matriz_dist, 0)

fig = go.Figure(data=go.Heatmap(
    z=matriz_dist,
    x=atomos,
    y=atomos,
    colorscale='RdYlBu_r',
    text=np.round(matriz_dist, 2),
    texttemplate='%{text} Ã…',
    textfont={"size": 10},
    colorbar=dict(title="Distancia (Ã…)")
))

fig.update_layout(
    title='Matriz de Distancias InteratÃ³micas',
    xaxis_title='Ãtomo',
    yaxis_title='Ãtomo',
    height=600,
    width=700
)

fig.show()
````

## ğŸ¯ AplicaciÃ³n Completa: Dashboard de AnÃ¡lisis Conformacional

Vamos a crear un anÃ¡lisis visual completo de un estudio conformacional.


````python
# Generar dataset de conformeros
np.random.seed(100)
n_conformeros = 100

df_conformeros = pd.DataFrame({
    'ID': [f'Conf_{i:03d}' for i in range(1, n_conformeros+1)],
    'Energia_kcal': np.random.exponential(2, n_conformeros),
    'Dipolo_D': np.random.gamma(2, 0.8, n_conformeros),
    'HOMO_eV': np.random.normal(-5.8, 0.4, n_conformeros),
    'LUMO_eV': np.random.normal(-1.2, 0.3, n_conformeros),
    'Volumen_A3': np.random.normal(250, 40, n_conformeros),
    'Area_A2': np.random.normal(350, 50, n_conformeros)
})

df_conformeros['Gap_eV'] = df_conformeros['LUMO_eV'] - df_conformeros['HOMO_eV']
df_conformeros['Energia_rel'] = df_conformeros['Energia_kcal'] - df_conformeros['Energia_kcal'].min()

print(f"Dataset generado: {len(df_conformeros)} conformeros")
print(df_conformeros.head())
````

````python
# Crear dashboard con mÃºltiples visualizaciones
fig = plt.figure(figsize=(16, 12))
gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)

# 1. DistribuciÃ³n de energÃ­as
ax1 = fig.add_subplot(gs[0, :2])
ax1.hist(df_conformeros['Energia_rel'], bins=25, color='steelblue', 
         edgecolor='black', alpha=0.7)
ax1.axvline(df_conformeros['Energia_rel'].mean(), color='red', 
           linestyle='--', linewidth=2, label=f'Media: {df_conformeros["Energia_rel"].mean():.2f}')
ax1.set_xlabel('EnergÃ­a relativa (kcal/mol)', fontsize=11)
ax1.set_ylabel('Frecuencia', fontsize=11)
ax1.set_title('A) DistribuciÃ³n de EnergÃ­as Conformacionales', fontsize=12, fontweight='bold', loc='left')
ax1.legend()
ax1.grid(True, alpha=0.3)

# 2. EstadÃ­sticas clave
ax2 = fig.add_subplot(gs[0, 2])
ax2.axis('off')
stats_text = f"""
ESTADÃSTICAS GLOBALES
{'='*25}

N conformeros: {len(df_conformeros)}

EnergÃ­a (kcal/mol):
  Min: {df_conformeros['Energia_rel'].min():.2f}
  Max: {df_conformeros['Energia_rel'].max():.2f}
  Media: {df_conformeros['Energia_rel'].mean():.2f}
  Std: {df_conformeros['Energia_rel'].std():.2f}

Gap HOMO-LUMO (eV):
  Media: {df_conformeros['Gap_eV'].mean():.2f}
  Std: {df_conformeros['Gap_eV'].std():.2f}

Momento Dipolar (D):
  Media: {df_conformeros['Dipolo_D'].mean():.2f}
  Max: {df_conformeros['Dipolo_D'].max():.2f}
"""
ax2.text(0.1, 0.9, stats_text, transform=ax2.transAxes, 
        fontsize=9, verticalalignment='top', fontfamily='monospace',
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

# 3. EnergÃ­a vs Momento Dipolar
ax3 = fig.add_subplot(gs[1, 0])
scatter = ax3.scatter(df_conformeros['Energia_rel'], df_conformeros['Dipolo_D'],
                     c=df_conformeros['Gap_eV'], cmap='viridis', 
                     s=60, alpha=0.6, edgecolors='black', linewidth=0.5)
ax3.set_xlabel('EnergÃ­a relativa (kcal/mol)', fontsize=10)
ax3.set_ylabel('Momento dipolar (Debye)', fontsize=10)
ax3.set_title('B) EnergÃ­a vs Momento Dipolar', fontsize=11, fontweight='bold', loc='left')
cbar = plt.colorbar(scatter, ax=ax3)
cbar.set_label('Gap (eV)', fontsize=9)
ax3.grid(True, alpha=0.3)

# 4. Gap HOMO-LUMO
ax4 = fig.add_subplot(gs[1, 1])
ax4.scatter(df_conformeros['HOMO_eV'], df_conformeros['LUMO_eV'],
           c=df_conformeros['Energia_rel'], cmap='RdYlGn_r',
           s=60, alpha=0.6, edgecolors='black', linewidth=0.5)
ax4.set_xlabel('HOMO (eV)', fontsize=10)
ax4.set_ylabel('LUMO (eV)', fontsize=10)
ax4.set_title('C) Niveles de Frontera', fontsize=11, fontweight='bold', loc='left')
ax4.grid(True, alpha=0.3)

# 5. Volumen vs Ãrea
ax5 = fig.add_subplot(gs[1, 2])
ax5.scatter(df_conformeros['Volumen_A3'], df_conformeros['Area_A2'],
           c=df_conformeros['Energia_rel'], cmap='coolwarm',
           s=60, alpha=0.6, edgecolors='black', linewidth=0.5)
ax5.set_xlabel('Volumen (Å²)', fontsize=10)
ax5.set_ylabel('Ãrea superficial (Å²)', fontsize=10)
ax5.set_title('D) Propiedades GeomÃ©tricas', fontsize=11, fontweight='bold', loc='left')
ax5.grid(True, alpha=0.3)

# 6. Boxplots comparativos
ax6 = fig.add_subplot(gs[2, :])
# Clasificar en grupos
df_conformeros['Grupo'] = pd.cut(df_conformeros['Energia_rel'], 
                                 bins=[0, 2, 5, 100],
                                 labels=['Baja (0-2)', 'Media (2-5)', 'Alta (>5)'])

grupos = df_conformeros.groupby('Grupo')
box_data = [group['Gap_eV'].values for name, group in grupos]
positions = range(1, len(box_data)+1)

bp = ax6.boxplot(box_data, positions=positions, widths=0.6, patch_artist=True,
                boxprops=dict(facecolor='lightblue', alpha=0.7),
                medianprops=dict(color='red', linewidth=2),
                whiskerprops=dict(linewidth=1.5),
                capprops=dict(linewidth=1.5))

ax6.set_xticklabels(['Baja\n(0-2 kcal/mol)', 'Media\n(2-5 kcal/mol)', 'Alta\n(>5 kcal/mol)'])
ax6.set_xlabel('ClasificaciÃ³n por EnergÃ­a', fontsize=11)
ax6.set_ylabel('Gap HOMO-LUMO (eV)', fontsize=11)
ax6.set_title('E) Gap HOMO-LUMO por Rango de EnergÃ­a', fontsize=12, fontweight='bold', loc='left')
ax6.grid(True, alpha=0.3, axis='y')

# TÃ­tulo general
fig.suptitle('ANÃLISIS CONFORMACIONAL COMPLETO', 
            fontsize=18, fontweight='bold', y=0.995)

plt.show()

print("\nğŸ“Š Dashboard generado con 5 visualizaciones complementarias")
````

## ğŸ’¡ Buenas PrÃ¡cticas de VisualizaciÃ³n

### Principios Fundamentales

1. **Claridad sobre complejidad**
   - GrÃ¡fico simple y directo > grÃ¡fico elaborado confuso
   - Elimina elementos innecesarios (chartjunk)

2. **Etiquetas apropiadas**
   - Ejes siempre con unidades
   - TÃ­tulos descriptivos
   - Leyendas claras

3. **Colores significativos**
   - Usa colores consistentes
   - Considera daltonismo (evita rojo-verde)
   - Mapas de color apropiados

4. **RelaciÃ³n datos-tinta**
   - Maximiza informaciÃ³n por pixel
   - Evita decoraciÃ³n excesiva

### Checklist para Figuras CientÃ­ficas

```
âœ… TÃ­tulo descriptivo
âœ… Etiquetas de ejes con unidades
âœ… TamaÃ±o de fuente legible (â‰¥10 pt)
âœ… Leyenda cuando hay mÃºltiples series
âœ… Grid sutil (alpha < 0.5)
âœ… ResoluciÃ³n apropiada (â‰¥300 dpi)
âœ… Formato vectorial para publicaciÃ³n
âœ… Colores distinguibles
âœ… Sin distorsiÃ³n de datos
âœ… Barras de error cuando aplique
```

### Errores Comunes a Evitar

âŒ Ejes sin unidades  
âŒ GrÃ¡ficos 3D innecesarios  
âŒ Colores sin significado  
âŒ Escala inapropiada (no empezar en 0 cuando es engaÃ±oso)  
âŒ Demasiada informaciÃ³n en un grÃ¡fico  
âŒ Fuentes muy pequeÃ±as  
âŒ GrÃ¡ficos de torta con >5 categorÃ­as
