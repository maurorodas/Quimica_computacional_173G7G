## üìö ¬øQu√© es una Consola o Terminal?

Una **consola** (tambi√©n llamada **terminal** o **l√≠nea de comandos**) es una interfaz de texto que permite interactuar con el sistema operativo mediante comandos escritos. 

### ¬øPor qu√© es importante en Qu√≠mica Computacional?

- üî¨ Muchos programas de qu√≠mica computacional se ejecutan desde la l√≠nea de comandos
- üìä Permite automatizar tareas repetitivas mediante scripts
- üíª Facilita el trabajo en servidores y clusters de c√°lculo
- üöÄ Proporciona mayor control y eficiencia que interfaces gr√°ficas

## 2Ô∏è‚É£ Navegaci√≥n B√°sica

### üìç `pwd` - ¬øD√≥nde estoy?

El comando `pwd` (Print Working Directory) muestra el directorio actual:

````python
# Mostrar el directorio actual
!pwd
````

### üìÇ `ls` - ¬øQu√© hay aqu√≠?

El comando `ls` (list) muestra el contenido del directorio actual:

````python
# Listar archivos y directorios
!ls
````

````python
# Listar con m√°s detalles (permisos, tama√±o, fecha)
!ls -lh
````

````python
# Listar incluyendo archivos ocultos
!ls -la
````

**Opciones comunes de `ls`:**
- `-l`: formato largo (detalles)
- `-a`: mostrar archivos ocultos
- `-h`: tama√±os legibles para humanos (KB, MB, GB)
- `-t`: ordenar por fecha de modificaci√≥n
- `-r`: orden inverso

## 3Ô∏è‚É£ Gesti√≥n de Directorios

### üìÅ `mkdir` - Crear Directorios

Creemos una estructura de directorios para organizar un proyecto de qu√≠mica computacional:

````python
# Crear un directorio principal para nuestro proyecto
!mkdir proyecto_quimica

# Crear subdirectorios
!mkdir proyecto_quimica/datos
!mkdir proyecto_quimica/resultados
!mkdir proyecto_quimica/scripts

# Verificar la estructura creada
!ls -R proyecto_quimica
````

### üö∂ `cd` - Cambiar de Directorio

El comando `cd` (change directory) permite movernos entre directorios:

**Nota:** En Jupyter Notebook, `cd` es un comando m√°gico y no requiere `!`

````python
# Cambiar al directorio del proyecto
%cd proyecto_quimica

# Verificar d√≥nde estamos
!pwd
````

````python
# Cambiar a un subdirectorio
%cd datos
!pwd
````

````python
# Subir un nivel (directorio padre)
%cd ..
!pwd
````

````python
# Volver al directorio home
%cd ~
!pwd
````

**Rutas importantes:**
- `.` = directorio actual
- `..` = directorio padre (un nivel arriba)
- `~` = directorio home del usuario
- `/` = directorio ra√≠z del sistema

## 4Ô∏è‚É£ Gesti√≥n de Archivos

### üìù Crear Archivos

Podemos crear archivos de varias formas:

````python
# Volver al directorio del proyecto
%cd /content/proyecto_quimica

# Crear un archivo vac√≠o con touch
!touch datos/experimento1.txt

# Crear un archivo con contenido usando echo
!echo "Datos de experimento 1" > datos/experimento1.txt

# Verificar que se cre√≥
!ls -lh datos/
````

````python
%%bash
# Crear un archivo de coordenadas moleculares (XYZ)
cat > moleculas/organicas/metano.xyz << EOF
5
Metano - CH4
C   0.000   0.000   0.000
H   0.631   0.631   0.631
H  -0.631  -0.631   0.631
H  -0.631   0.631  -0.631
H   0.631  -0.631  -0.631
EOF

# Ver el contenido
cat moleculas/organicas/metano.xyz
````

### üëÅÔ∏è Ver Contenido de Archivos

Existen varios comandos para visualizar el contenido de archivos:

````python
# cat: muestra todo el contenido del archivo
!cat moleculas/organicas/metano.xyz
````

````python
# head: muestra las primeras l√≠neas (por defecto 10)
!head -n 3 moleculas/organicas/metano.xyz
````

````python
# tail: muestra las √∫ltimas l√≠neas
!tail -n 3 moleculas/organicas/metano.xyz
````

````python
# wc: contar l√≠neas, palabras y caracteres
!wc moleculas/organicas/metano.xyz
````

### üìã Copiar Archivos - `cp`

````python
# Copiar un archivo
!cp moleculas/organicas/metano.xyz moleculas/organicas/metano_backup.xyz

# Copiar con otro nombre
!cp moleculas/organicas/metano.xyz resultados/metano_optimizado.xyz

# Verificar
!ls -lh moleculas/organicas/
!ls -lh resultados/
````

````python
# Copiar un directorio completo con -r (recursivo)
!cp -r moleculas moleculas_backup

# Verificar
!ls -R moleculas_backup
````

### ‚úÇÔ∏è Mover/Renombrar Archivos - `mv`

````python
# Renombrar un archivo
!mv resultados/metano_optimizado.xyz resultados/CH4_optimizado.xyz

# Mover un archivo a otro directorio
!mv moleculas/organicas/metano_backup.xyz datos/

# Verificar
!ls resultados/
!ls datos/
````

### üóëÔ∏è Eliminar Archivos y Directorios - `rm`

‚ö†Ô∏è **CUIDADO:** El comando `rm` elimina archivos de forma permanente. No hay papelera de reciclaje.

````python
# Eliminar un archivo
!rm datos/metano_backup.xyz

# Eliminar un directorio vac√≠o con rmdir
!rmdir moleculas/inorganicas

# Eliminar un directorio con contenido con rm -r
!rm -r moleculas_backup

# Verificar
!ls -R moleculas/
````

## 5Ô∏è‚É£ Comandos √ötiles Adicionales

### üîç `grep` - Buscar Texto en Archivos

````python
# Buscar l√≠neas que contienen "H" (hidr√≥geno)
!grep "H" moleculas/organicas/metano.xyz
````

````python
# Buscar con n√∫mero de l√≠nea
!grep -n "H" moleculas/organicas/metano.xyz
````

### üîó `|` - Pipes (Tuber√≠as)

Los pipes permiten conectar la salida de un comando con la entrada de otro:

````python
# Contar cu√°ntos √°tomos de hidr√≥geno hay
!grep "H" moleculas/organicas/metano.xyz | wc -l
````

````python
# Listar archivos y contar cu√°ntos hay
!ls -l | wc -l
````

### üîé `find` - Buscar Archivos

El comando `find` busca archivos y directorios por nombre, tipo, tama√±o, etc.:

````python
# Buscar todos los archivos .xyz
!find . -name "*.xyz"
````

````python
# Buscar todos los directorios
!find . -type d
````
