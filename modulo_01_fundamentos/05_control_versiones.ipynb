## ğŸ“š Â¿QuÃ© es el Control de Versiones?

El **control de versiones** es un sistema que registra los cambios en archivos a lo largo del tiempo, permitiÃ©ndote recuperar versiones especÃ­ficas mÃ¡s adelante.

### Â¿Por quÃ© es crucial en QuÃ­mica Computacional?

- ğŸ“ **Trazabilidad**: Saber exactamente quÃ© cambios se hicieron y cuÃ¡ndo
- ğŸ”¬ **Reproducibilidad**: Recuperar la versiÃ³n exacta del cÃ³digo que generÃ³ resultados publicados
- ğŸ¤ **ColaboraciÃ³n**: Trabajar simultÃ¡neamente sin sobrescribir el trabajo de otros
- ğŸ”™ **RecuperaciÃ³n**: Volver a versiones anteriores cuando algo falla
- ğŸ“Š **DocumentaciÃ³n**: Historial completo del desarrollo del proyecto
- ğŸŒ **Compartir**: Publicar cÃ³digo y datos de forma organizada

## ğŸ”‘ Conceptos Fundamentales de Git

### Repositorio (Repository)
Carpeta de proyecto que Git rastrea. Contiene:
- Todos los archivos del proyecto
- Historial completo de cambios
- ConfiguraciÃ³n de Git

### Commit
InstantÃ¡nea (snapshot) del proyecto en un momento especÃ­fico.
- Captura el estado de todos los archivos
- Incluye un mensaje descriptivo
- Tiene un identificador Ãºnico (hash)

### Rama (Branch)
LÃ­nea de desarrollo independiente.
- `main` o `master`: rama principal
- `develop`: desarrollo activo
- `feature/nueva-funcionalidad`: caracterÃ­sticas especÃ­ficas
- `fix/corregir-bug`: correcciones

### Staging Area (Ãrea de PreparaciÃ³n)
Zona intermedia donde preparas los cambios antes de confirmarlos.

### Flujo de Trabajo BÃ¡sico
```
Directorio de trabajo â†’ Staging Area â†’ Repositorio local â†’ Repositorio remoto
      (editar)            (add)         (commit)           (push)
```

## ğŸ“ Agregar CÃ³digo: Script de CÃ¡lculo Molecular

### Crear Script de Python


````python
# Crear script para calcular masa molecular
script_content = '''#!/usr/bin/env python3
"""
Script para calcular la masa molecular de compuestos quÃ­micos.
"""

# Masas atÃ³micas (uma)
MASAS_ATOMICAS = {
    'H': 1.008, 'C': 12.011, 'N': 14.007, 'O': 15.999,
    'S': 32.065, 'P': 30.974, 'Cl': 35.453, 'Br': 79.904
}

def calcular_masa_molecular(formula):
    """
    Calcula la masa molecular de una fÃ³rmula quÃ­mica simple.
    
    Ejemplo: H2O -> 18.015 uma
    """
    masa_total = 0.0
    i = 0
    
    while i < len(formula):
        elemento = formula[i]
        i += 1
        
        # Leer nÃºmero si existe
        numero = ""
        while i < len(formula) and formula[i].isdigit():
            numero += formula[i]
            i += 1
        
        cantidad = int(numero) if numero else 1
        
        if elemento in MASAS_ATOMICAS:
            masa_total += MASAS_ATOMICAS[elemento] * cantidad
        else:
            print(f"âš ï¸ Advertencia: Elemento '{elemento}' no reconocido")
    
    return masa_total

if __name__ == "__main__":
    # Ejemplos de uso
    compuestos = ["H2O", "CO2", "C6H12O6", "H2SO4"]
    
    print("Calculadora de Masa Molecular\\n")
    print("-" * 40)
    
    for compuesto in compuestos:
        masa = calcular_masa_molecular(compuesto)
        print(f"{compuesto:10s} = {masa:.3f} uma")
'''

with open("scripts/masa_molecular.py", "w") as f:
    f.write(script_content)

print("âœ… Script creado: scripts/masa_molecular.py")
!cat scripts/masa_molecular.py
````

## ğŸŒ¿ Trabajar con Ramas (Branches)

Las ramas permiten desarrollar caracterÃ­sticas sin afectar el cÃ³digo principal.

### Caso de Uso: Agregar Nueva Funcionalidad

Queremos agregar una funciÃ³n para calcular volumen molar, pero sin romper el cÃ³digo funcional.


````python
# Ver ramas existentes
print("ğŸŒ³ Ramas actuales:\n")
!git branch

# Crear nueva rama
!git branch feature/volumen-molar

# Ver ramas nuevamente
print("\nğŸŒ³ DespuÃ©s de crear nueva rama:\n")
!git branch
````

````python
# Cambiar a la nueva rama
!git checkout feature/volumen-molar

# Verificar rama actual
print("âœ… Rama actual:")
!git branch
````

### Desarrollar en la Nueva Rama


````python
# Agregar nueva funciÃ³n al script
nueva_funcion = '''

def calcular_volumen_molar(masa_molecular, densidad):
    """
    Calcula el volumen molar de un compuesto.
    
    Args:
        masa_molecular (float): Masa molecular en g/mol
        densidad (float): Densidad en g/cmÂ³
    
    Returns:
        float: Volumen molar en cmÂ³/mol
    """
    return masa_molecular / densidad

def calcular_densidad(masa_molecular, volumen_molar):
    """
    Calcula la densidad a partir del volumen molar.
    
    Args:
        masa_molecular (float): Masa molecular en g/mol
        volumen_molar (float): Volumen molar en cmÂ³/mol
    
    Returns:
        float: Densidad en g/cmÂ³
    """
    return masa_molecular / volumen_molar
'''

# Agregar al final del script (antes del if __name__)
with open("scripts/masa_molecular.py", "r") as f:
    contenido = f.read()

# Insertar antes de if __name__
partes = contenido.split('if __name__ == "__main__":')
contenido_nuevo = partes[0] + nueva_funcion + '\n\nif __name__ == "__main__":' + partes[1]

with open("scripts/masa_molecular.py", "w") as f:
    f.write(contenido_nuevo)

print("âœ… Nuevas funciones agregadas")
````

````python
# Commit en la rama de desarrollo
!git add scripts/masa_molecular.py
!git commit -m "âœ¨ Agregar funciones de volumen molar y densidad"

# Ver historial con ramas
print("\nğŸ“œ Historial con ramas:\n")
!git log --oneline --graph --all --decorate
````

### Fusionar (Merge) la Rama

Una vez que la caracterÃ­stica estÃ¡ completa y probada, la fusionamos con `main`.


````python
# Volver a la rama main
!git checkout main

# Fusionar feature/volumen-molar en main
!git merge feature/volumen-molar -m "ğŸ”€ Merge: Integrar funciones de volumen molar"

# Ver historial
print("\nğŸ“œ Historial despuÃ©s del merge:\n")
!git log --oneline --graph --all
````

````python
# Eliminar rama ya fusionada (opcional)
!git branch -d feature/volumen-molar

print("\nğŸŒ³ Ramas restantes:")
!git branch
````
