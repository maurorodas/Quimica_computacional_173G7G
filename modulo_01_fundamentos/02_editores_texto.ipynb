## üìö ¬øQu√© es un Editor de Texto Plano?

Un **editor de texto plano** es un programa que permite crear y editar archivos de texto sin formato (sin negritas, colores o estilos como en Word).

### ¬øPor qu√© son importantes en Qu√≠mica Computacional?

- üìÑ Los archivos de entrada para programas qu√≠micos son de texto plano
- üîß Scripts de automatizaci√≥n se escriben en texto plano
- üíæ Archivos de configuraci√≥n y par√°metros son texto plano
- üìä Datos de salida frecuentemente son texto plano
- üöÄ M√°s eficientes y ligeros que procesadores de texto

### Texto Plano vs Texto Enriquecido

| Texto Plano | Texto Enriquecido |
|-------------|-------------------|
| ‚úÖ `.txt`, `.py`, `.sh`, `.xyz` | ‚ùå `.docx`, `.pdf`, `.rtf` |
| ‚úÖ Sin formato especial | ‚ùå Con formato (negritas, fuentes) |
| ‚úÖ Archivos peque√±os | ‚ùå Archivos grandes |
| ‚úÖ Editable en cualquier sistema | ‚ùå Requiere software espec√≠fico |

## 2Ô∏è‚É£ Vim/Vi - El Editor Poderoso

**Vim** (Vi Improved) es un editor muy potente y eficiente, aunque tiene una curva de aprendizaje m√°s pronunciada.

### ‚ö†Ô∏è Nota Importante
**Vim es un editor interactivo modal.** Al igual que nano, debes practicarlo en tu **terminal local**. Aqu√≠ aprender√°s los conceptos fundamentales.

### ¬øPor qu√© aprender vim?

- üöÄ Extremadamente r√°pido y eficiente
- üí™ Disponible en casi todos los sistemas Unix/Linux
- ‚å®Ô∏è Permite editar sin usar el mouse
- üîß Altamente configurable y extensible
- üìö Muy usado en servidores remotos

### Modos de Vim

Vim tiene varios **modos** de operaci√≥n:

1. **Modo Normal** (predeterminado): Para navegar y ejecutar comandos
2. **Modo Inserci√≥n**: Para escribir texto
3. **Modo Visual**: Para seleccionar texto
4. **Modo Comando**: Para ejecutar comandos del editor

### Comandos Esenciales de Vim

#### Abrir y Salir
| Comando | Acci√≥n |
|---------|--------|
| `vim archivo.txt` | Abrir archivo |
| `:q` | Salir (si no hay cambios) |
| `:q!` | Salir sin guardar cambios |
| `:w` | Guardar |
| `:wq` o `ZZ` | Guardar y salir |

#### Modo Inserci√≥n
| Comando | Acci√≥n |
|---------|--------|
| `i` | Insertar antes del cursor |
| `a` | Insertar despu√©s del cursor |
| `I` | Insertar al inicio de la l√≠nea |
| `A` | Insertar al final de la l√≠nea |
| `o` | Nueva l√≠nea debajo |
| `O` | Nueva l√≠nea arriba |
| `Esc` | Volver al modo normal |

#### Navegaci√≥n (Modo Normal)
| Comando | Acci√≥n |
|---------|--------|
| `h, j, k, l` | Izquierda, abajo, arriba, derecha |
| `w` | Siguiente palabra |
| `b` | Palabra anterior |
| `0` | Inicio de l√≠nea |
| `$` | Final de l√≠nea |
| `gg` | Inicio del archivo |
| `G` | Final del archivo |

#### Edici√≥n (Modo Normal)
| Comando | Acci√≥n |
|---------|--------|
| `x` | Borrar car√°cter |
| `dd` | Borrar l√≠nea |
| `yy` | Copiar l√≠nea |
| `p` | Pegar |
| `u` | Deshacer |
| `Ctrl+r` | Rehacer |
| `/texto` | Buscar texto |
| `:%s/old/new/g` | Reemplazar (todo el archivo) |

### Ejemplo 1: Crear un Archivo de Input para GAMESS

````python
# Crear un archivo de input b√°sico para GAMESS
!cat > h2o_gamess.inp << 'EOF'
 $CONTRL SCFTYP=RHF RUNTYP=ENERGY $END
 $BASIS GBASIS=STO NGAUSS=3 $END
 $DATA
Water molecule - RHF/STO-3G
C1
O     8.0   0.0   0.0   0.119262
H     1.0   0.0   0.763239  -0.477047
H     1.0   0.0  -0.763239  -0.477047
 $END
EOF

print("Archivo de input de GAMESS creado:")
!cat h2o_gamess.inp
````

### Ejemplo 2: Script de Bash para Automatizaci√≥n

````python
# Crear un script para procesar m√∫ltiples archivos
!cat > procesar_moleculas.sh << 'EOF'
#!/bin/bash
# Script para procesar archivos XYZ

echo "Procesando archivos XYZ..."

for archivo in *.xyz
do
    echo "Procesando: $archivo"
    # Contar el n√∫mero de √°tomos
    atoms=$(head -n 1 "$archivo")
    echo "  N√∫mero de √°tomos: $atoms"
done

echo "Proceso completado."
EOF

# Dar permisos de ejecuci√≥n
!chmod +x procesar_moleculas.sh

# Mostrar el script
!cat procesar_moleculas.sh
````

````python
# Ejecutar el script
!./procesar_moleculas.sh
````

### Ejemplo 3: Archivo de Par√°metros de Din√°mica Molecular

````python
# Crear un archivo de par√°metros para GROMACS
!cat > em.mdp << 'EOF'
; Archivo de minimizaci√≥n de energ√≠a para GROMACS
; Par√°metros de minimizaci√≥n
integrator  = steep         ; Algoritmo de descenso m√°s pronunciado
emtol       = 1000.0        ; Tolerancia de fuerza (kJ/mol/nm)
emstep      = 0.01          ; Paso inicial (nm)
nsteps      = 50000         ; N√∫mero m√°ximo de pasos

; Par√°metros de salida
nstlog      = 100           ; Frecuencia de escritura en log
nstenergy   = 100           ; Frecuencia de escritura de energ√≠as

; Par√°metros de vecinos
cutoff-scheme   = Verlet
nstlist         = 10
ns_type         = grid
rlist           = 1.0

; Interacciones electrost√°ticas
coulombtype     = PME
rcoulomb        = 1.0

; Interacciones de van der Waals
vdwtype         = Cut-off
rvdw            = 1.0

; Condiciones de contorno
pbc             = xyz       ; Peri√≥dicas en todas direcciones
EOF

print("Archivo de par√°metros MD creado:")
!head -n 15 em.mdp
````

## 5Ô∏è‚É£ Herramientas √ötiles para Edici√≥n

### sed - Editor de Flujo

`sed` permite editar archivos sin abrirlos interactivamente:

````python
# Crear un archivo de ejemplo
!echo -e "Temperatura: 300\nPresi√≥n: 1.0\nTiempo: 1000" > parametros.txt
print("Archivo original:")
!cat parametros.txt
````

````python
# Reemplazar temperatura con sed
!sed -i 's/300/310/g' parametros.txt
print("\nArchivo modificado:")
!cat parametros.txt
````

### awk - Procesamiento de Texto

`awk` es excelente para procesar archivos tabulares:

````python
# Extraer solo las coordenadas de un archivo XYZ (ignorando las primeras 2 l√≠neas)
print("Coordenadas de agua:")
!awk 'NR>2 {print $2, $3, $4}' agua.xyz
````
